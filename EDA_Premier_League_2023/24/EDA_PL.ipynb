```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("premier-player-23-24.csv")
## Se carga la base de datos de excel para tener acceso a la informacion y se procede con el analisis
```

```python
df.head()
## Se muestran las primeras 5 filas correspondientes a los primeros 5 jugadores registrados en la base de datos
## para dar un pequeño vistazo a las estadisticas registradas (goles, asistencias, partidos, etc...)
```

```python
df.tail()
## De igual manera se muestran los ultimos 5 jugadores registrados
```

```python
df.describe()
## Se realiza una descripcion basica de los datos mostrando un 
# analisis descriptivo de los mismos (conteo,media,desviacion y cuartiles)
```

```python
goles_assist = df.nlargest(10, 'G+A')[['Player', 'G+A', 'Gls', 'Ast']]
goles_assist

# Se muestran los 10 jugadores con más contribuciines de goles para reflejar su influencia en el juego
# la estadistica G+A resulta de concatenar las cifras de goles y asistencias y juntarlas en una sola cifra

## Resalta el jugador Cole Palmer como lider en este rubro, aunque no es lider en goles o asistencias por separado
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='G+A', y='Player', data=goles_assist, palette='viridis')
plt.title('Jugadores Con Mayores Contribuciones de Gol')
plt.xlabel('Goles + Asistencias')
plt.ylabel('Jugador')
plt.show()

# Se reflejan con una grafica los jugadores mas influyentes mostrados en la tabla anterior 
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='xG', y='Gls', data=df, hue='Pos', palette='deep')
plt.title('Goles Esperados (xG) vs. Goles Anotados')
plt.xlabel('Goles Esperados (xG)')
plt.ylabel('Goles Anotados (Gls)')
plt.show()

## Esta grafica busca contrastar los goles anotados con los goles que segun una ponderacion calculada por la liga
## pudo haber anotado el jugador para dar un vistazo acerca de las oportunidades desperdiciadas por estos jugadores

## Posiciones: 
## MF: Mediocampista
## GK: Portero
## DF: Defensa
## FW: Delantero 

## Aquellos datos que indican dos posiciones son para aquellos jugadores que se desempeñan en dos posiciones,
## expresando asi su polivalencia en el juego 
```

```python
top_avances = df.nlargest(10, 'PrgC')[['Player', 'PrgC']]
top_avances

plt.figure(figsize=(10, 6))
sns.barplot(x='PrgC', y='Player', data=top_avances, palette='magma')
plt.title('Top 10 Jugadores con Avances Progresivos')
plt.xlabel('Avances Progresivos')
plt.ylabel('Jugador')
plt.show()

## Se muestran los 10 jugadores con mas avances progresivos en la liga, aquellos que ayudan con una carrera, un pase 
## hacia adelante o una conduccion de balon a salir a su equipo de la presion e iniciar un ataque prometedor 

## Jeremy Doku, extremo del Manchester City es el jugador más explosivo a la hora iniciar ataques, sirviendo muchas veces 
## como solucion ante la presion de los rivales 

## Cabe destacar que el colombiano Luis Diaz se encuentra en la septima posicion en este rubro 
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribucion de Jugadores por Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

## Esta grafica muestra la distribucion de frecuencias de los jugadores segun su edad, mostrando que la mayoria de 
## jugadores de la liga se encuentran entre los 20 y 25 años de edad, lo cual va de la mano con la alta intensidad 
## y la rapidez en los partidos en la liga y evidenciando la alta inversion que realizan los equipos de la liga en
## jugadores jovenes para poder competir en condiciones  

## Se puede ver que la distribucion de los datos sigue una distribucion semejante a la normal
```

```python
pos_stats = df.groupby('Pos')[['Gls', 'Ast']].sum().reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x='Pos', y='Gls', data=pos_stats, color='coral', label='Goles')
sns.barplot(x='Pos', y='Ast', data=pos_stats, color='lightblue', label='Asistencias', alpha=0.7)
plt.title('Goles y Asistencias por Posicion')
plt.xlabel('Posicion')
plt.ylabel('Total')
plt.legend()
plt.show()

## Aqui se evidencian las contribuciones de gol por posicion en la liga. Como se espera aquellos jugadores que aportan
## mas a este rubro son ofensivos. Ademas, cabe resaltar las elevadas contribuciones de parte de los defensas, los cuales
## son parte crucial en la creacion de juego, especialmente los laterales los cuales se incorporan al ataque lanzando
## centros y haciendo pases clave
```

```python
pais_stats = df.groupby('Nation')[['Gls', 'Ast']].sum().reset_index()

top_paises = pais_stats.nlargest(10, 'Gls')

plt.figure(figsize=(14, 8))
sns.barplot(x='Gls', y='Nation', data=top_paises, color='lightgreen', label='Goles')
sns.barplot(x='Ast', y='Nation', data=top_paises, color='lightcoral', label='Asistencias', alpha=0.7)
plt.title('Top 10 Paises en Goles y Asistencias')
plt.xlabel('Total')
plt.ylabel('Pais')
plt.legend()
plt.show()

## En los paises con mayores contribuciones de gol se expresa que la liga esta dominada por jugadores ingleses
## mostrando que se valora y se apuesta por el mercado local. Jugadores como Ollie Watkins o Dominic Solanke son 
## grandes exponentes de la liga y son ingleses, al igual que Cole Palmer el cual como se menciono anteriormente 
## es lider en G+A 
```

```python
goleador_equipo = df.loc[df.groupby('Team')['Gls'].idxmax()][['Team', 'Player', 'Gls']]
goleador_equipo = goleador_equipo.sort_values(by='Gls', ascending=False).reset_index(drop=True)
print(goleador_equipo)

## Se muestra la tabla de goleadores de cada equipo, donde en 7 de los 20 equipos su goleador es ingles 
## en 2 es Noruego y el unico goleador asiatico es Heung-Min Son, el coreano cuenta con 17 anotaciones
```

```python
df['Min_per_Gls'] = df['Min'] / df['Gls']
df['Min_per_Gls'] = df['Min_per_Gls'].replace([float('inf'), float('nan')], 0)
most_min_per_goal = df.nlargest(10, 'Min_per_Gls')[['Player', 'Min', 'Gls', 'Min_per_Gls']]
most_min_per_goal

## Aqui se muestran los 10 jugadores que necesitan mas minutos en cancha para anotar un gol, destacando que esta no
## es su faceta principal, pues la gran mayoria de estos son defensores 
```

# Conclusiones 

- En la Premier League, a diferencia de lo que se cree, se apuesta por el mercado local, pues sus jugadores mas influyentes son generalmente ingleses como lo reflejo la base de datos. 

- Es un aspecto a mejorar por parte de los jugadores la tasa de conversion de goles, pues la grafica que enfrenta los goles anotados con los esperados expresa las oportunidades desperdiciadas y es un punto de partida para mejorar.

- Sorprendentemente las posiciones defensivas son clave en la creacion de juego en los equipos ingleses, pues generalmente estos equipos adoptan posturas agresivas y lanzan sus laterales al ataque y buscan pases incisivos de parte de sus centrales los cuales en ocasiones terminan en gol

- Aunque influyen en el juego generalmente repartiendo asistencias, la posicion en el campo que menos goles anota en promedio son las defensas ya que estos generalmente tienen otra funcion en el juego.

- Los equipos de la liga apuestan fuertemente por jugadores jovenes entre los 20 y 25 años ya que estos se adaptan facilmente al ritmo de la liga y en promedio son menos propensos a sufrir lesiones.
